```python
#data slicer
indmin = data.groupby(data.index.to_period('M')).apply(lambda x: x.index.min()).values
indmax = data.groupby(data.index.to_period('M')).apply(lambda x: x.index.max()).values

#itg lists with frames
periods = []
cors = []

p_periods = []

for i in range(0, 61):
    #data slicer
    period = data.loc[indmin[i]:indmax[i+11]]
    period_corr = period.corr()
    
    cors.append(period_corr.values.astype(np.float32))
    periods.append(period.values.astype(np.float32))
    
    #choose 10 assets:max corr, 10 assets:min corr, 5 assets:randomly
    abs_corr = period_corr.abs().sum().sort_values()

    asset_labels = []
    asset_labels += abs_corr[:10].index.to_list()
    asset_labels += abs_corr.index[random.sample(range(11, len(data.columns)-10), 5)].to_list()
    asset_labels += abs_corr[-10:].index.to_list()

    p_periods.append(period[asset_labels].values.astype(np.float32))
```

```python
from tqdm import tqdm
from IPython.display import clear_output
```

```python
from copulae import ClaytonCopula, GumbelCopula
```

```python
def clayton_param(df):
    cc = ClaytonCopula(dim=2)
    cc.fit(df)
    cc_ll = cc.log_lik(df)
    cc_p = cc.params
    clear_output(wait=True)
    return cc_ll, cc_p

def gumbel_param(df):
    gc = GumbelCopula(dim=2)
    gc.fit(df)
    gc_ll = gc.log_lik(df)
    gc_p = gc.params
    clear_output(wait=True)
    return gc_ll, gc_p
```

```python
copula_cors = []

for period in tqdm(periods):
    corr_matrix = np.zeros((200, 200))
    for i in range(200):
        corr_matrix[i, i] = 1.0
        print("ASSET"+str(i))
        for j in range(i+1, 200):
            pre = pd.DataFrame([period.T[i], period.T[j]]).T
            try:
                cc_ll, cc_p = clayton_param(pre)
            except:
                cc_ll = 1
                cc_p = 1
            try:
                gc_ll, gc_p = gumbel_param(pre)
            except:
                gc_ll = 1
                gc_p = 1
            if gc_ll >= cc_ll:
                corr_matrix[i][j] = np.sin( np.pi * (1 - 1 / gc_p) / 2)
                corr_matrix[j][i] = np.sin( np.pi * (1 - 1 / gc_p) / 2)
            else:
                corr_matrix[i][j] = np.sin(np.pi * (cc_p / (cc_p + 2)) / 2)
                corr_matrix[j][i] = np.sin(np.pi * (cc_p / (cc_p + 2)) / 2)
            print("COLUMN"+str(j))
        clear_output(wait=True)
                
    copula_cors.append(corr_matrix)
```

```python
for i in range(len(copula_cors)):
    np.savetxt('intermediate_data/archimedian/archimedian_cors%s.gz' %i
               , copula_cors[i]
               , delimiter=',')
```
