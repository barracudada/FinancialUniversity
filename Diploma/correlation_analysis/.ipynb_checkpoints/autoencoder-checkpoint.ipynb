```python
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_excel('./data/data.xlsx')
data['Unnamed: 0'] = pd.to_datetime(data['Unnamed: 0'])
data = data.set_index('Unnamed: 0')
data.index.name = None
data.head()
```

```python
pre = data.groupby([data.index.year, data.index.month]).agg(["first", "last"])

data_pct_m = pd.DataFrame(index = data.groupby(data.index.to_period('M')).apply(lambda x: x.index.max()).values
                     , columns = data.columns)
for i in data.columns:
    data_pct_m[i] = ( (pre[i]['last'] - pre[i]['first']) / pre[i]['first'] ).values

print(data_pct_m.shape)
data_pct_m.head()
```

```python
data_pct = data.pct_change().dropna()

pdata = data[:366]
pdata_pct = pdata.pct_change().dropna()
pdata_pct_m = data_pct_m[:12]
```

```python
from scipy import stats
import random
```

```python
def make_data_uniform(df):
    us = []
    for col in df.columns:
        X = stats.Uniform(a=min(df[col].values), b=max(df[col].values))
        us.append(X.cdf(df[col]))
    us = pd.DataFrame(us).T
    us.columns = df.columns
    return us

#data slicer
indmin = data.groupby(data.index.to_period('M')).apply(lambda x: x.index.min()).values
indmax = data.groupby(data.index.to_period('M')).apply(lambda x: x.index.max()).values

#itg lists with frames
dfs_train = []
dfs_test = []
dfs_test_labels = []
cors = []

for i in range(0, 61):
    #data slicer
    pre = data.loc[indmin[i]:indmax[i+11]]
    pre_corr = pre.corr()
    
    #slice corr
    cors.append(pre_corr.values.astype(np.float32))
    
    #make data uniform 
    dfs_train.append(make_data_uniform(pre).values.astype(np.float32))
    
    #choose 10 assets:max corr, 10 assets:min corr, 5 assets:randomly
    abs_corr = pre_corr.abs().sum().sort_values()

    asset_labels = []
    asset_labels += abs_corr[:5].index.to_list()
    asset_labels += abs_corr.index[random.sample(range(11, len(data.columns)-10), 15)].to_list()
    asset_labels += abs_corr[-5:].index.to_list()
    dfs_test_labels.append(asset_labels)

    dfs_test.append(make_data_uniform(pre[asset_labels]).values.astype(np.float32))
```

```python
from sklearn.model_selection import train_test_split
```

```python
#prices
x_train, x_val, y_train, y_val = train_test_split(dfs_train, dfs_test, test_size=0.3, shuffle=False)
#correlations
cor_train, cor_val = train_test_split(cors, test_size=0.3, shuffle=False)
```

```python
len(x_train), len(x_val), len(y_train), len(y_val), len(cor_train), len(cor_val)
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
x_train_loader = torch.utils.data.DataLoader(x_train, batch_size=1)
x_val_loader = torch.utils.data.DataLoader(x_val, batch_size=1)
y_train_loader = torch.utils.data.DataLoader(y_train, batch_size=1)
y_val_loader = torch.utils.data.DataLoader(y_val, batch_size=1)

cor_train_loader = torch.utils.data.DataLoader(cor_train, batch_size=1)
cor_val_loader = torch.utils.data.DataLoader(cor_val, batch_size=1)
```

```python
strd = 1
    
class part_encoder(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.encoder = nn.Sequential(
            nn.Conv2d(1, 32, kernel_size=(1,26), stride=strd),
            nn.ReLU(),
            nn.Conv2d(32, 64, kernel_size=(1,26), stride=strd),
            nn.ReLU(), 
            nn.Conv2d(64, 128, kernel_size=(1,26), stride=strd),
            nn.ReLU(),
            nn.Conv2d(128, 128, kernel_size=(1,26), stride=strd),
            nn.ReLU(),
            nn.Conv2d(128, 64, kernel_size=(1,26), stride=strd),
            nn.ReLU(),
            nn.Conv2d(64, 32, kernel_size=(1,26), stride=strd),
            nn.ReLU(),
            nn.Conv2d(32, 1, kernel_size=(1,26), stride=strd),
        )
    def forward(self, sample):
        reconstructed = self.encoder(sample)
        return reconstructed
    
class part_decoder(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.decoder = nn.Sequential(
            nn.ConvTranspose2d(1, 32, kernel_size=(26,26), stride=strd),
            nn.ReLU(),
            nn.ConvTranspose2d(32, 64, kernel_size=(26,26), stride=strd),
            nn.ReLU(), 
            nn.ConvTranspose2d(64, 128, kernel_size=(26,26), stride=strd),
            nn.ReLU(),
            nn.ConvTranspose2d(128, 128, kernel_size=(26,26), stride=strd),
            nn.ReLU(),
            nn.Dropout(0.2),
            nn.ConvTranspose2d(128, 64, kernel_size=(26,26), stride=strd),
            nn.ReLU(),
            nn.ConvTranspose2d(64, 32, kernel_size=(26,26), stride=strd),
            nn.ReLU(),
            nn.ConvTranspose2d(32, 1, kernel_size=(26,26), stride=strd),
        )
    def forward(self, sample):
        return self.decoder(sample)
```

```python
from tqdm import tqdm
```

```python
n_ep = 10
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'

#encoder
model_encoder = part_encoder().to(device)
optimizer = torch.optim.Adam(model_encoder.parameters(), lr=0.001)
loss_fn = F.mse_loss

train_encoder_losses = []
val_encoder_losses = []

X_encoded_train_list = []
X_encoded_val_list = []

for epoch in tqdm(range(n_ep)):
    #encoder train
    model_encoder.train()
    train_encoder_losses_per_epoch = []
    for index, (X, Y) in enumerate(zip(x_train_loader, y_train_loader)):
        optimizer.zero_grad()
        X_encoded = model_encoder(X.to(device))
        #save rec
        X_encoded_train_list.append(X_encoded)
        #save loss
        loss = loss_fn(X_encoded, Y.to(device))
        loss.backward()
        optimizer.step()
        train_encoder_losses_per_epoch.append(loss.item())
    train_encoder_losses.append(np.mean(train_encoder_losses_per_epoch))

    #encoder valuation
    model_encoder.eval()
    val_encoder_losses_per_epoch = []
    with torch.no_grad():
        for index, (X, Y) in enumerate(zip(x_val_loader, y_val_loader)):
            X_encoded = model_encoder(X.to(device))
            #save rec
            X_encoded_val_list.append(X_encoded)
            #save loss
            loss = loss_fn(X_encoded, Y.to(device))
            val_encoder_losses_per_epoch.append(loss.item())
            
    val_encoder_losses.append(np.mean(val_encoder_losses_per_epoch))
```

```python
torch.save(model_encoder.state_dict(), 'model_encoder.pth')
```

```python
import copulae
from copulae import GaussianCopula, StudentCopula

def copula_eval(values):
    values = values.detach().numpy()[0][0]
    
    gc = GaussianCopula(dim=values.shape[1])
    gc.fit(values)
    gc_ll = gc.log_lik(values)
    gc_corr = gc.sigma
    return torch.from_numpy(np.array([gc_corr]).astype(np.float32))
    #sc = StudentCopula(dim=values.shape[1])
    #sc.fit(values)
    #sc_ll = sc.log_lik(values)
    #sc_corr = sc.sigma
    
    #if gc_ll >= sc_ll:
    #    return torch.from_numpy(np.array([gc_corr]).astype(np.float32)) 
    #else:
    #    return torch.from_numpy(np.array([sc_corr]).astype(np.float32)) 
```

```python
from IPython.display import clear_output

n_ep = 10
device = 'cuda:0' if torch.cuda.is_available() else 'cpu'

#decoder
model_decoder = part_decoder().to(device)
optimizer = torch.optim.Adam(model_decoder.parameters(), lr=0.00001)
train_decoder_losses = []
val_decoder_losses = []

loss_fn = F.mse_loss

X_encoded_train_copula_list = []
X_encoded_train_copula_large_list = []
X_encoded_val_copula_list = []
X_encoded_val_copula_large_list = []

for epoch in tqdm(range(n_ep)):
    #decoder train
    model_decoder.train()
    train_decoder_losses_per_epoch = []
    for index, (X, Y) in enumerate(zip(X_encoded_train_list_loader, cor_train_loader)):
        optimizer.zero_grad()
        #compute copula cor
        X_encoded_copula = copula_eval(X)
        X_encoded_train_copula_list.append(X_encoded_copula)
        #decode
        X_encoded_copula_large = model_decoder(X_encoded_copula.to(device))
        X_encoded_train_copula_large_list.append(X_encoded_copula_large)
        #save loss
        loss = loss_fn(X_encoded_copula_large, Y.to(device))
        loss.backward()
        optimizer.step()
        train_decoder_losses_per_epoch.append(loss.item())
        print('train '+str(index))
    train_decoder_losses.append(np.mean(train_decoder_losses_per_epoch))

    #encoder valuation
    model_decoder.eval()
    val_decoder_losses_per_epoch = []
    with torch.no_grad():
        for index, (X, Y) in enumerate(zip(X_encoded_val_list_loader, cor_val_loader)):
            #compute copula
            X_encoded_copula = copula_eval(X)
            X_encoded_val_copula_list.append(X_encoded_copula)
            #decode
            X_encoded_copula_large = model_decoder(X_encoded_copula.to(device))
            X_encoded_val_copula_large_list.append(X_encoded_copula_large)
            #save loss
            loss = loss_fn(X_encoded_copula_large, Y.to(device))
            val_decoder_losses_per_epoch.append(loss.item())
            print('train '+str(index))
    val_decoder_losses.append(np.mean(val_decoder_losses_per_epoch))
    clear_output(wait=True)
```

```python
fig, ax = plt.subplots()

ax.plot(range(n_ep), train_decoder_losses, label='train')
ax.plot(range(n_ep), val_decoder_losses, label='valid')
ax.set_title('Обучение декодировщика')
ax.set_xlabel('Эпоха обучения',  fontsize = 10)
ax.set_ylabel('Средняя квадратичная ошибка', fontsize = 10)
plt.legend(['train', 'valid'])
plt.show()
```
