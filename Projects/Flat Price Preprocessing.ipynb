### 3.  Удаление и импутация
Удаляем квартиры без общей площади

```python
df = df.dropna(subset=['Площадь.общая'])
```

Импутируем пропущенное время до метро медианным значением

```python
median_value=df['Время.до.метро'].median()
df['Время.до.метро'].fillna(value=median_value, inplace=True)
```

Импутация в разрезе групп:

```python
df['Площадь.кухни'].fillna(df.groupby('Комнатность')['Площадь.кухни'].transform('mean'))
```

Кодирование категориальных переменных

```python

one_hot = ['Новостройка', 'Метро.Район', 'Источник.получения']
one_hot_features=pd.get_dummies(df[one_hot])
one_hot_features
```

Соберем таблицу

```python
## Если в таблице нет лишних полей, можно написать:
#df=df.drop(columns=one_hot)
#df=df.join(one_hot_features)

## Или же собираем ее "вручную"
df=df[['Комнатность', 'Время.до.метро', 'Этаж', 'Этажность', 'Площадь.общая']].join(one_hot_features).join(df['Ценавруб.'])
```
