```python
from pycoingecko import CoinGeckoAPI
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup as bs
```

```python
cg = CoinGeckoAPI()
```

```python
cg.get_price(ids='bitcoin', vs_currencies='usd')
```

```python
df_coins = pd.DataFrame.from_dict(cg.get_coins_list(), orient='columns')[['id', 'name', 'symbol']]
df_coins
```

```python
coin = df_coins['id']
URL1 = "https://coinmarketcap.com/currencies/"
URL2 = "/holders/"
URL_final = URL1 + coin + URL2 
links = pd.DataFrame(URL_final)
```

```python
for i in range(1000):
    a = requests.get(links.iat[i, 0])
    b = a.status_code
    df_coins.at[i, 'Status'] = b

df_coins.head(8)
```

200 - получен ответ от сервера (монета нашлась)
404 - страница не найдена
403 - превышен лимит запросов
